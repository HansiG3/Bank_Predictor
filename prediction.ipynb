```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
model = load_model('model.h5')
with open('onehot_encoder_geography.pkl', 'rb') as file:
    onehot_encoder_geo = pickle.load(file)
with open('label_encoder_gender.pkl', 'rb') as file:
    label_encoder_gender = pickle.load(file)
with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)

```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python

# Convert to DataFrame
input_df = pd.DataFrame([input_data])

# Encode Gender (label encoding)
input_df['Gender'] = label_encoder_gender.transform(
    input_df['Gender']
)

# Encode Geography (one-hot encoding) â€“ use DataFrame to avoid warning
geo_encoded = onehot_encoder_geo.transform(
    pd.DataFrame({'Geography': input_df['Geography']})
).toarray()

geo_encoded_df = pd.DataFrame(
    geo_encoded,
    columns=onehot_encoder_geo.get_feature_names_out(['Geography'])
)

# Drop Geography column
input_df.drop('Geography', axis=1, inplace=True)

# Concatenate encoded geography
input_df = pd.concat([input_df, geo_encoded_df], axis=1)

# Scale input
input_scaled = scaler.transform(input_df)

# Predict
prediction = model.predict(input_scaled)

print("Prediction probability:", prediction[0][0])
```
